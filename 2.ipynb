```python
!gdown --id 1shBZAw8s7b2jse4rmD8RhJ1WF-5zxfB7

import numpy as np
import pandas as pd
column_names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATTO", "B", "LSTAT", "MEDV"]
df = pd.read_csv("housing.csv", names=column_names, delimiter=r"\s+", header=None)
df.head
```

```python
type(df)
```

```python
df.describe()
```

```python
df.groupby("CHAS")["MEDV"].sum()
```

```python
df.groupby(["CHAS", "RAD"])["MEDV"].sum()
```

```python
df.groupby(["CHAS", "RAD"])["MEDV"].sum().reset_index()
```

```python
import seaborn as sns

data = df.groupby(["CHAS", "RAD"])["MEDV"].mean().reset_index()
sns.barplot(x="RAD", y="MEDV", hue="CHAS", data=data)
```

```python
data = df.groupby(["CHAS", "RAD"])["MEDV"].sum().reset_index()
sns.barplot(x="RAD", y="MEDV", hue="CHAS", data=data)
```

```python
data = df.groupby(["CHAS", "RAD"])["MEDV"].mean().reset_index()
data.plot(kind="bar", stacked=True)
```
