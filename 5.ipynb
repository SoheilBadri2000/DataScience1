```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define function
def sigmoid(t):
  return 1/(1+np.exp(-t))
```

```python
sigmoid(0)
```

```python
sigmoid(np.inf)
```

```python
sigmoid(-np.inf)
```

```python
x = np.linspace(-10, 10, 100)
y = sigmoid(x)
y
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, lw=5)
```

```python
plt.plot(x, y, linestyle="--")
plt.xticks(rotation=45)
```
